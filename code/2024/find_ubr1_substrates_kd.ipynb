## Load Libraries

```r
library('tidyverse')
library('readxl')
```

Load data

```r
df <- read_excel("../../data/2024/aaw4912_timms_sm_data-file-s3.xlsx", sheet = 1, skip = 1) %>%
    janitor:: clean_names() %>%
    rename(bin1_aavs = bin1_3) %>%
    print()
```

```r
gene_df <- read_excel("../../data/2024/aaw4912_timms_sm_data-file-s3.xlsx", sheet = 1, range = "B2:B21000") %>%
    janitor::clean_names() %>%
    print()
control_df <- read_excel("../../data/2024/aaw4912_timms_sm_data-file-s3.xlsx", sheet = 1, range = "C2:I21000") %>%
    janitor::clean_names() %>%
    rename_all(~ paste0("aavs1_", .))%>%
    print()


ubrko1_df <- read_excel("../../data/2024/aaw4912_timms_sm_data-file-s3.xlsx", sheet = 1, range = "K2:Q21000") %>%
    janitor::clean_names() %>%
    rename_all(~ paste0("clone1_", .))%>%
    print()

    
ubrko2_df <- read_excel("../../data/2024/aaw4912_timms_sm_data-file-s3.xlsx", sheet = 1, range = "S2:Y21000") %>%
    janitor::clean_names() %>%
    rename_all(~ paste0("clone2_", .))%>%
    print()
```

```r
joined_df <- gene_df %>%
    bind_cols(control_df, ubrko1_df, ubrko2_df) %>%
    write_csv("../../data/2024/ubr1_substrates.csv") %>%
    print(n = 400)
    
```
